```python
df = pd.read_csv('Cars Datasets 2025.csv', encoding='latin-1')
df
```

```python
df_cleaned = df.copy()

# Set exchange rate
euro_to_usd = 1.1

def clean_price(price):
    if pd.isna(price):
        return None

    price_str = str(price).strip()

    # Remove "Math input error"
    price_str = price_str.replace("Math input error", "").strip()

    # Normalize dash characters
    price_str = price_str.replace("–", "-").replace("—", "-")

    # Handle Euro (e.g., €33,000)
    if "€" in price_str:
        match = re.search(r"€\s*([\d,]+)", price_str)
        if match:
            value = int(match.group(1).replace(",", ""))
            return round(value * euro_to_usd)

    # Handle price range (e.g., "12,000 - 15,000")
    range_match = re.match(r"([\d,]+)\s*-\s*([\d,]+)", price_str)
    if range_match:
        low = int(range_match.group(1).replace(",", ""))
        high = int(range_match.group(2).replace(",", ""))
        return (low + high) // 2

    # Handle standalone number (with optional $)
    num_match = re.search(r"([\d,]+)", price_str)
    if num_match:
        return int(num_match.group(1).replace(",", ""))

    return None

# Apply cleaner
df_cleaned["Cars Prices"] = df_cleaned["Cars Prices"].apply(clean_price)

# Rename the column
df_cleaned.rename(columns={"Cars Prices": "Cars Prices (USD)"}, inplace=True)
```

```python
# Normalize Fuel Types column
fuel_col = df_cleaned["Fuel Types"].str.lower().fillna("")

# Include anything that contains "petrol" or "gas"
is_petrol_or_gas = fuel_col.str.contains("petrol") | fuel_col.str.contains("gas")

# Exclude if it also contains "hybrid", "electric", or "plug"
not_mixed = ~fuel_col.str.contains("hybrid|electric|plug")

# Final filter for pure gas cars
pure_gas_mask = is_petrol_or_gas & not_mixed

# Create a new DataFrame of just gas cars
gas_cars = df_cleaned[pure_gas_mask].copy()

# Optional: Rename their fuel type to "Gas"
gas_cars["Fuel Types"] = "Gas"
```

```python
group1 = gas_cars[gas_cars['Fuel Types'] == 'Gas']['Cars Prices (USD)'].dropna()
group2 = hybrid_cars[hybrid_cars['Fuel Types'] == 'Hybrid']['Cars Prices (USD)'].dropna()

t_stat, p_val = ttest_ind(group1, group2, equal_var=False)

print("t-statistic:", t_stat)
print("p-value:", p_val)
```

# Since p-value is greater than alpha we fail to reject the null hypothesis at 0.05 significance level. There is no statistically significant difference in the average prices of gas cars and hybrid cars in your dataset. The observed difference in means could just be due to random chance.



```python
def clean_torque(val):
    if pd.isna(val):
        return None
    
    # Ensure string, strip whitespace, remove 'Nm' and extra spaces
    val = str(val).replace("Nm", "").strip()
    
    # Normalize dashes and spaces (e.g., '100 - 140', '100–140', '190 210')
    val = re.sub(r"[–\-]", "-", val)  # replace dash variants with hyphen
    numbers = re.findall(r"\d+", val)
    
    # Convert to numbers
    numbers = [int(n) for n in numbers]
    
    if len(numbers) == 1:
        return numbers[0]
    elif len(numbers) >= 2:
        return sum(numbers[:2]) / 2  # average of first two numbers
    else:
        return None

# Apply to your dataframe
hybrid_cars["Torque (Nm)"] = hybrid_cars["Torque"].apply(clean_torque)

# Drop or keep the original column
hybrid_cars = hybrid_cars.drop(columns=["Torque"])
```

```python
def clean_torque(val):
    if pd.isna(val):
        return None
    
    val = str(val).replace("Nm", "").strip()
    val = re.sub(r"[–\-]", "-", val)  # normalize dashes
    numbers = re.findall(r"\d+", val)
    numbers = [int(n) for n in numbers]

    if len(numbers) == 1:
        return numbers[0]
    elif len(numbers) >= 2:
        return sum(numbers[:2]) / 2  # average of first two
    else:
        return None

# Apply it to gas_cars
gas_cars["Torque (Nm)"] = gas_cars["Torque"].apply(clean_torque)

# Drop original column if no longer needed
gas_cars = gas_cars.drop(columns=["Torque"])
```

```python
def clean_and_convert(performance_str):
    if pd.isna(performance_str):
        return None

    # Normalize weird dashes
    normalized = str(performance_str)
    normalized = re.sub(r'[–—‑−]', '-', normalized)
    normalized = normalized.replace('to', '-')

    # Extract numbers
    numbers = re.findall(r'\d+\.?\d*', normalized)

    if len(numbers) == 2:
        avg_kmh = (float(numbers[0]) + float(numbers[1])) / 2
    elif len(numbers) == 1:
        avg_kmh = float(numbers[0])
    else:
        return None

    # Convert to 0–60 MPH
    return avg_kmh / 1.036
```

```python
hybrid_cars["Performance (0 - 60 MPH)"] = hybrid_cars["Performance_kmh"].apply(clean_and_convert)
```

```python
# Successes and sample sizes
x1, n1 = 220, 898  # Gas_Cars
x2, n2 = 23, 114  # Hybrid_Cars

# Sample proportions
p1 = x1 / n1
p2 = x2 / n2

# Pooled proportion (under H0)
p_pool = (x1 + x2) / (n1 + n2)

# Standard error
se = np.sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))

# Z-score
z = (p1 - p2) / se

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

print(f"Z-statistic: {z:.4f}")
print(f"P-value: {p_value:.4f}")
```

# Since p-value = 0.3087 is greater than any common alpha level (like 0.05 or 0.01), you fail to reject the null hypothesis. Conclusion: You do not have statistically significant evidence to say that the proportion of high-performance gas cars is different from that of hybrid cars.
