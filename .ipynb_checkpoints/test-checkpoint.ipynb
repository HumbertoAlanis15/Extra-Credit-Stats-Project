```python
df_cleaned = df.copy()

# Set exchange rate
euro_to_usd = 1.1

def clean_price(price):
    if pd.isna(price):
        return None

    price_str = str(price).strip()

    # Remove "Math input error"
    price_str = price_str.replace("Math input error", "").strip()

    # Normalize dash characters
    price_str = price_str.replace("–", "-").replace("—", "-")

    # Handle Euro (e.g., €33,000)
    if "€" in price_str:
        match = re.search(r"€\s*([\d,]+)", price_str)
        if match:
            value = int(match.group(1).replace(",", ""))
            return round(value * euro_to_usd)

    # Handle price range (e.g., "12,000 - 15,000")
    range_match = re.match(r"([\d,]+)\s*-\s*([\d,]+)", price_str)
    if range_match:
        low = int(range_match.group(1).replace(",", ""))
        high = int(range_match.group(2).replace(",", ""))
        return (low + high) // 2

    # Handle standalone number (with optional $)
    num_match = re.search(r"([\d,]+)", price_str)
    if num_match:
        return int(num_match.group(1).replace(",", ""))

    return None

# Apply cleaner
df_cleaned["Cars Prices"] = df_cleaned["Cars Prices"].apply(clean_price)

# Rename the column
df_cleaned.rename(columns={"Cars Prices": "Cars Prices (USD)"}, inplace=True)
```

# # Since p-value is greater than alpha we fail to reject the null hypothesis at 0.05 significance level. There is no statistically significant difference in the mean torque between gas and hybrid cars. The data do not provide evidence to suggest that the average torque differs between the two fuel types.

```python
gas_torque = np.random.normal(loc=412, scale=242, size=898)
gas_fuel_type = ['Gas'] * 898

# Example data for Hybrid cars (assume lower torque)
hybrid_torque = np.random.normal(loc=424, scale=209, size=114) # mean 424, std 209
hybrid_fuel_type = ['Hybrid'] * 114

# Combine into a DataFrame
df = pd.DataFrame({
    'Fuel Type': gas_fuel_type + hybrid_fuel_type,
    'Torque (lb-ft)': np.concatenate([gas_torque, hybrid_torque])
})
```

```python
group_stats = df.groupby('Fuel Type')['Torque (lb-ft)'].agg(
    mean_torque='mean',
    sem_torque=lambda x: x.sem(), # Calculate Standard Error of the Mean
    count='count'
)
```

```python
group_stats['ci_lower'] = group_stats['mean_torque'] - 1.96 * group_stats['sem_torque']
group_stats['ci_upper'] = group_stats['mean_torque'] + 1.96 * group_stats['sem_torque']

print("Group Statistics:")
print(group_stats)
print("\n" + "="*30 + "\n")
```

```python
torque_gas = df[df['Fuel Type'] == 'Gas']['Torque (lb-ft)']
torque_hybrid = df[df['Fuel Type'] == 'Hybrid']['Torque (lb-ft)']
```

```python
t_stat, p_value = stats.ttest_ind(torque_gas, torque_hybrid, equal_var=False)

print(f"Two-Sample Independent T-Test Results (Gas vs. Hybrid Torque):")
print(f"  T-statistic: {t_stat:.3f}")
print(f"  P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print(f"  Result: Reject the null hypothesis (p < {alpha}). There is a statistically significant difference in average torque between Gas and Hybrid cars.")
else:
    print(f"  Result: Fail to reject the null hypothesis (p >= {alpha}). There is no statistically significant difference in average torque between Gas and Hybrid cars.")

print("\n" + "="*30 + "\n")
```

```python
plt.figure(figsize=(8, 6))

# Plot the means as points
plt.errorbar(
    x=group_stats.index,
    y=group_stats['mean_torque'],
    yerr=group_stats['sem_torque'], # This is the +/- error for SEM
    fmt='o', # Plot as circles
    capsize=5, # Size of the error bar caps
    color='blue',
    ecolor='red', # Color of the error bars
    elinewidth=2, # Thickness of error lines
    markersize=10, # Size of the mean markers
    label='Mean Torque (with SEM error bars)'
)

plt.title('Average Torque Comparison: Gas vs. Hybrid Cars')
plt.xlabel('Fuel Type')
plt.ylabel('Average Torque (lb-ft)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
def clean_and_convert(performance_str):
    if pd.isna(performance_str):
        return None

    # Normalize weird dashes
    normalized = str(performance_str)
    normalized = re.sub(r'[–—‑−]', '-', normalized)
    normalized = normalized.replace('to', '-')

    # Extract numbers
    numbers = re.findall(r'\d+\.?\d*', normalized)

    if len(numbers) == 2:
        avg_kmh = (float(numbers[0]) + float(numbers[1])) / 2
    elif len(numbers) == 1:
        avg_kmh = float(numbers[0])
    else:
        return None

    # Convert to 0–60 MPH
    return avg_kmh / 1.036
```

```python
# Successes and sample sizes
x1, n1 = 220, 898  # Gas_Cars
x2, n2 = 23, 114  # Hybrid_Cars

# Sample proportions
p1 = x1 / n1
p2 = x2 / n2

# Pooled proportion (under H0)
p_pool = (x1 + x2) / (n1 + n2)

# Standard error
se = np.sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))

# Z-score
z = (p1 - p2) / se

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

print(f"Z-statistic: {z:.4f}")
print(f"P-value: {p_value:.4f}")
```

# # Since p-value = 0.3087 is greater than any common alpha level (like 0.05 or 0.01), you fail to reject the null hypothesis. Conclusion: You do not have statistically significant evidence to say that the proportion of high-performance gas cars is different from that of hybrid cars.

```python
data = {
    'Group': ['Gas Cars', 'Hybrid Cars'],
    'Successes': [220, 23],
    'Failures': [678, 91 ],   
    'Total': [898, 114]     
}
df = pd.DataFrame(data)
```

```python
df['Prop_Success'] = df['Successes'] / df['Total']
df['Prop_Failure'] = df['Failures'] / df['Total']

print("DataFrame with Proportions:\n", df)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

bar_width = 0.6 # Adjust as needed

bars1 = ax.bar(df['Group'], df['Prop_Success'], bar_width, label='Successes', color='#66c2a5') # A nice green shade
bars2 = ax.bar(df['Group'], df['Prop_Failure'], bar_width, bottom=df['Prop_Success'], label='Failures', color='#fc8d62') # A nice orange shade

ax.set_ylabel('Proportion')
ax.set_title('Proportion of Successes and Failures by Group')
ax.set_ylim(0, 1) # Proportions always go from 0 to 1
ax.legend(title='Outcome')

def add_labels(bars, proportions, is_top=False):
    for bar, prop in zip(bars, proportions):
        yval = bar.get_height() / 2 + (bar.get_y() if is_top else 0) # Center the text
        ax.text(bar.get_x() + bar.get_width()/2, yval,
                f'{prop:.1%}', ha='center', va='center', color='black', fontsize=9)

add_labels(bars1, df['Prop_Success'])
add_labels(bars2, df['Prop_Failure'], is_top=True)

plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7) # Add a grid for better readability
plt.show()
```
